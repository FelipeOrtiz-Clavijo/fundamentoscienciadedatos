# **2. Cargar conjunto de datos**

Aquí se carga el conjunto de datos de flores Iris utilizando la función load_dataset() de **Seaborn** y se guarda en un DataFrame de pandas llamado iris_df. La función display() de Python se utiliza para mostrar las primeras y últimas filas del conjunto de datos. Esta parte es necesaria para acceder y analizar los datos en el código.

**sns.load_dataset("iris"):** La función load_dataset() de Seaborn se utiliza para cargar el conjunto de datos de flores Iris. El resultado es un DataFrame de Pandas que contiene las muestras de las flores Iris.

**display():** Esta función se utiliza para mostrar las primeras y últimas filas del DataFrame.

```python
iris_df = sns.load_dataset("iris")
display(iris_df.head(),iris_df.tail())
```

En esta parte se describe el problema que se quiere resolver.

El [conjunto de datos de la flor Iris](https://es.wikipedia.org/wiki/Conjunto_de_datos_flor_iris) contiene 150 muestras de las medidas del ancho y el largo del pétalo y del sépalo para las tres especies de esta flor (Iris setosa, Iris virginica e Iris versicolor). A partir de la muestra de datos se desean responder la siguiente pregunta:



La pregunta planteada es: **¿Cuál es la longitud y ancho típicos del pétalo y del sépalo para cada uno de los tipos de flores?**


# **4. Transformación de datos para facilitar la visualización**

En esta parte del código, se crea una copia del DataFrame iris_df llamada iris_melt. Luego, se utiliza la función pd.melt() para reestructurar el DataFrame en un formato más adecuado para la visualización de datos.

**La transformación de datos es necesaria para facilitar el análisis y la creación de gráficos.**

**iris_df.copy():** Se crea una copia del DataFrame original para evitar modificarlo directamente.

**pd.melt():** Esta función de Pandas se utiliza para transformar el DataFrame de un formato ancho a uno largo. Esto facilita la visualización de los datos en gráficos de dispersión. La función toma como argumentos el DataFrame, la columna que se utilizará como identificador (species), el nombre de la nueva columna que contendrá los nombres de las variables originales (Variables) y el nombre de la nueva columna que contendrá los valores correspondientes a esas variables (Values).

```python
iris_melt = iris_df.copy()
iris_melt = pd.melt(
    iris_melt,
    id_vars="species",
    var_name="variables",
    value_name="values",
)
iris_melt.head()
```

# **5. Visualización de datos utilizando gráficos de dispersión**

Esta parte del código utiliza la función sns.catplot() para crear gráficos de dispersión de "swarm" que permiten comparar las variables del conjunto de datos por especie. La visualización de datos es esencial para explorar y entender las características y patrones en los datos.



```python
sns.catplot(
    x="variables",
    y="values",
    data=iris_melt,
    col="species",
    kind="swarm",
)
plt.show()
```

**Visualización de datos utilizando gráficos de dispersión por variables y especies**

Se utiliza la función sns.catplot() para crear gráficos de dispersión de "swarm" que permiten comparar las especies de Iris por cada variable. Esta visualización de datos es útil para identificar diferencias y similitudes entre las especies con respecto a las medidas del pétalo y el sépalo.

```python
sns.catplot(
    x="species",
    y="values",
    data=iris_melt,
    col="variables",
    kind="swarm",
    col_wrap=4,
)
plt.show()
```


**Nota grafica**

Este fragmento de código crea gráficos de dispersión tipo "swarm" agrupados por variables y especies utilizando la libreria Seaborn y, además, añade líneas horizontales para representar las medias y medianas de cada especie.

**sns.catplot():** Esta función de Seaborn crea un gráfico categórico con las siguientes opciones:

**x="species":** Define el eje x con la columna "species" del DataFrame iris_melt.

**y="values":** Define el eje y con la columna "values" del DataFrame iris_melt.

**data=iris_melt:** Especifica el DataFrame que se utilizará para el gráfico.

**col="variables":** Crea un gráfico separado para cada valor único en la columna "variables" del DataFrame.

**kind="swarm":** Define el tipo de gráfico categórico como "swarm", que es un gráfico de dispersión donde los puntos no se superponen.

**col_wrap=2:** Establece el número de columnas en la cuadrícula de gráficos a 2.

**colors =** ['tab:blue', 'tab:orange', 'tab:green']: Define una lista de colores para las líneas de media y mediana de cada especie.

**means = iris_df.groupby("species").mean():** Calcula la media de cada columna en el DataFrame iris_df agrupado por la columna "species".

**medians = iris_df.groupby("species").median():** Calcula la mediana de cada columna en el DataFrame iris_df agrupado por la columna "species".

El bucle for itera sobre los índices de los 4 ejes (cada eje corresponde a una variable) en los gráficos creados por sns.catplot():

**ax = g.axes[i_axes]:** Obtiene el objeto de eje actual en la iteración.
mean_values = means[means.columns[i_axes]]: Obtiene las medias de la variable correspondiente en la iteración.

**median_values = medians[medians.columns[i_axes]]:** Obtiene las medianas de la variable correspondiente en la iteración.

El siguiente bucle for itera sobre los colores, medias y medianas de las especies:

**g.axes[i_axes].axhline(mean_value, ls='--', color=color):** Dibuja una línea horizontal punteada en el valor de la media con el color correspondiente.

g**.axes[i_axes].axhline(median_value, ls=':', color=color):** Dibuja una línea horizontal de puntos en el valor de la mediana con el color correspondiente.

**plt.show():** Muestra los gráficos creados en una ventana separada.

```python
g = sns.catplot(
    x="species",
    y="values",
    data=iris_melt,
    col="variables",
    kind="swarm",
    col_wrap=2,
)

colors = ['tab:blue', 'tab:orange', 'tab:green']


means = iris_df.groupby("species").mean()
medians = iris_df.groupby("species").median()

for i_axes in range(4):
    ax = g.axes[i_axes]
    mean_values = means[means.columns[i_axes]]
    median_values = medians[medians.columns[i_axes]]
    for color, mean_value, median_value in zip(colors, mean_values, median_values):
        g.axes[i_axes].axhline(mean_value, ls='--', color=color)
        g.axes[i_axes].axhline(median_value, ls=':', color=color)



plt.show()
```

Aquí se calcula la moda para la variable sepal_length. Se utilizan funciones de las librerias **statistics** y **scipy** para realizar el cálculo y luego se muestra el resultado utilizando la función display() de Python.
